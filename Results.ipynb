Even though the final results are similar, it seems that a high number of unroll steps (40) makes the job of the critic very difficult and introduces a lot of flactuations, with no gain except maybe for the asymptotic result (I did not push the simulation more than strictly necessary). However an intermediate number of steps (e.g. 20) seems to be strictly superior in sample efficiency and final result than a low number of steps (e.g. 5 or 1). Basically the idea is that the signal flows to the previous states faster in the case of long number of unrollings (fast initial convergence), but also noisier (difficult final convergence).

A good thing to try would be to schedule a reduction of the number of steps used for the update as the episode proceedes.

Notice also that 20 here, that seems the best parameter, is just enough to cover the shortest path between every possible pair of points in a 10 x 10 grid (max distance is 18); so it could be that being also intrisically nearer to the dimension of the map somehow influences the result.

Note: each point is the average of the 500 preceding ones, so the curves are smoother and one can actually figure out the important trends better.

The $\tau$ coefficient regulates how faster is updated the critic's target network starting from the critic's parameters (that are updated after every trajectory). A "delayed" learning for the critic is claimed to make the learning more stable, but of course if the task is very simple it can be that it's just holding back the whole network progress. It should be also noted that the forgetting of the parameters at a certain time step t happens with an exponential rate, so even with high $\tau$ we have significant past information flowing. 

Probably the optimal coefficient depends also on the learning rate, because that is the main parameter that regulates both the speed of learning and the instability connected to large updates. In this particular case it seems that $\tau = 0.3$ is a good parameter.

```python
class OheNet(nn.Module):
    def __init__(self, map_size, k_in=3, k_out=24, n_features=32, pixel_hidden_dim=128, 
                 pixel_n_residuals=4, feature_hidden_dim=64, feature_n_residuals=4):
        
        super(OheNet, self).__init__()
        
        self.n_features = n_features
        
        self.OHE_conv = Convolution(k_in=k_in, k_out=k_out)
        self.pos_enc = PositionalEncoding(n_kernels=k_out, n_features=n_features)

        pixel_res_layers = nn.ModuleList([ResidualLayer(map_size**2, pixel_hidden_dim) 
                                          for _ in range(pixel_n_residuals)])
        self.pixel_res_block = nn.Sequential(*pixel_res_layers)

        self.maxpool = FeaturewiseMaxPool(pixel_axis=2)

        feature_res_layers = nn.ModuleList([ResidualLayer(n_features, feature_hidden_dim) 
                                            for _ in range(feature_n_residuals)])
        self.feature_res_block = nn.Sequential(*feature_res_layers)
```
